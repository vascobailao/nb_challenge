Part I - In this part, we will choose 500.000 observations of the original data in a random way.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_set = pd.read_csv("../data/train.gz")
```

```python
def convert_to_int(x):
    return int(x)
```

```python
train_set.id = train_set.id.apply(convert_to_int)
```

```python
train_set.head(10)
```

```python
train_set_indices = train_set.index.tolist()
```

```python
index_list_click = np.random.choice(train_set_indices, 500000, False).tolist()
```

```python
new_train = train_set.loc[ train_set.index.isin(index_list_click), : ]
```

```python
sns.countplot(new_train["click"])
```

```python
new_train.to_csv("../data/selected_data/small_train.gz", compression="gzip")
```
